```python
# for mearging the data frame
df_merge = pd.concat([df_fake, df_true], axis=0)
df_merge.head(10)
```

```python
# for removing the column
df = df_merge.drop(["title", "subject", "date"], axis=1)
df.head(10)
```

```python
#to suffle te data set
df = df.sample(frac=1)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
import re
import string

# remove unnecessary characters
def word_drop(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\\W', ' ', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```

```python
df.head(10)
```

```python
# To define an independent and dependent variable
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size = .25)
```

```python
# for vectorizing the x variable
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

# for importing logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR =LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
LR.score(xv_train, y_train)  # For training set accuracy
LR.score(xv_test, y_test)    # For test set accuracy

```

```python
pred_LR = LR.predict(xv_test)
```

### To print the actual predection value

```python
print(classification_report(y_test, pred_LR))
```
